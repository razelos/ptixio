```python
import pandas as pd
import os
```

```python
df = pd.read_csv("genius_songs_data.csv")
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df['pageviews'] = df['pageviews'].fillna(0)
```

```python
df['pageviews'] = df['pageviews'].astype(int)
```

```python
df['album'] = df['album'].fillna('Single/Unknown')
```

```python
df['featured_artists'] = df['featured_artists'].fillna('No feat')
```

```python
df['date'] = df['date'].fillna('no date given')
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
FILENAME = "clean_songs.csv"
# Check if file exists so we know if we need to write the header
file_exists = os.path.isfile(FILENAME)

df.to_csv(FILENAME, mode='w', index=False, header=not file_exists)
```
